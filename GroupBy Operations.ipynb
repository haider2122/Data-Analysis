```python
#Creating methods and using the apply method for specificity
```

```python
f_passenger=list(t.groupby(by='sex'))[0][1]# all female passengers
```

```python
def func(df):
    return df.mean() 
```

```python
func(f_passenger)#one way of executing method commands
```

```python
t.groupby('sex').apply(func)
```

```python
def fun1(g):
    return g[(g.survived==1)].nlargest(5,'age')
```

```python
t.groupby('sex').apply(fun1)#Oldest passenger in accordance with gorupby execution
```
