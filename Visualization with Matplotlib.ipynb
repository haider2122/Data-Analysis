```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
t=pd.read_csv('titanic.csv')
```

```python
t.plot()
plt.show()#basics, a default line graph appears
```

```python
t.plot(subplots=True,sharex=False,sharey=False,figsize=(8,8),title='Line chart')
plt.show()#setting our parameters
```

```python
t.age.plot()#a line graph only for age
plt.show()
```

```python
#Exploiting all parameters
t.age.plot(kind='line',c='blue',figsize=(12,5),fontsize=15,linestyle='-')
```

```python
plt.show()
```

```python
t.age.plot(kind='line',c='blue',figsize=(12,5),fontsize=15,linestyle='-',xlim=(0,50),ylim=(20,35))
#setting x and y limits
plt.show()
```

```python
xd=[]
for i in range(0,900,50):
    xd.append(i)
    
yd=[]

for i in range(0,80,2):
    yd.append(i)
#Creating new x and y distribution
```

```python
#Setting new distribution as xticks and yticks
#Setting new distribution as xticks and ytick
#Setting new distribution as xticks and yticks
t.age.plot(kind='line',c='blue',figsize=(50,50),fontsize=50,linestyle='-',xticks=xd,yticks=yd,rot=45)
```

```python
plt.show()
#new graph
#with rotating labels value with rot
```

```python
plt.title('Titanic')
plt.xlabel('passenger no',fontsize=10)
plt.ylabel('numeric age',fontsize=10)
plt.legend(loc='best')
plt.show()#Other important customization
```

```python
xt=[x for x in range(0,90,10)]
yt=[y for y in range(0,30,2)]#Creating histograms, cover doubts lec=92

x=(t.age.value_counts())
x.plot(kind='hist',figsize=(14,10),fontsize=15,x=xt,y=yt)
plt.show()#Bins can also be directly given through plot method
```

```python
#another way to creat a histogram
p=t.age.value_counts(dropna=True)
plt.hist(p,bins=10,cumulative=True,color='green',orientation="horizontal")
plt.show()#This another way to create a hist histograms and play with different paramets
```
