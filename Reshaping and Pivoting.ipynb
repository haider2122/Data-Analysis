```python
#using pd.crosstab
t.head(2)
```

```python
pd.crosstab(index=t.sex,columns=t.pclass)#Its forming relations between index and columns
#Can be used for multiple data frames simultaneously
```

```python
pd.crosstab(index=t.sex,columns=t.pclass,values=t.fare,aggfunc='sum')
#Performing changes based on specific values
```

```python
pd.crosstab(index=t.sex,columns=t.pclass,values=t.fare,aggfunc='sum',normalize=True,margins=True)
#additional functionality using normaling and margins
```

```python
pd.crosstab(index=t.sex,columns=t.pclass,values=t.survived,aggfunc='mean',margins=True)
```

```python
t.groupby(by=['sex','pclass']).survived.mean().unstack(-1)
#Similar results using groupby
#lacks margins and normalize and other addn functionality
```
