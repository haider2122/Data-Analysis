```python
import pandas as pd
```

```python
t=pd.read_csv('titanic.csv')
```

```python
t.info()
```

```python
#Simple Random Sampling
```

```python
t.sample(88)
```

```python
t.sample(88).groupby("pclass").fare.mean()
```

```python
#Systematic Sampling
```

```python
t.iloc[2:600:21].groupby('sex').age.mean()
```

```python
#Stratified Sampling
```

```python
pd.cut(t.age,[0,20,40,55,80],labels=['Child','Young Adult','Adult','Senior citizen']).value_counts()
```

```python
t['Age_cat']=pd.cut(t.age,[0,20,40,55,80],labels=['Child','Young Adult','Adult','Senior citizen'])
```

```python
t.groupby('Age_cat').survived.mean()
```
