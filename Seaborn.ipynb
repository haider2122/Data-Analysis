```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
t=pd.read_csv('titanic.csv')
```

```python
plt.figure(figsize=(6,3))
sns.countplot(data=t,x='sex')
plt.show()
```

```python
plt.figure(figsize=(6,3))
sns.set(font_scale=2,style='darkgrid',palette='viridis')
sns.countplot(data=t,x='sex',hue='pclass')
plt.show()
#Using hue to split male femal on x axis wrt to pclass
#Using set() for adding additional decorative features
```

```python
#Creating scatter plots with strip
plt.figure(figsize=(8,6))
sns.set(font_scale=2,palette='viridis')
sns.stripplot(data=t,x='sex',y='age',hue='pclass')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.set(font_scale=2,palette='viridis')
sns.stripplot(data=t,x='sex',y='age',hue='pclass',jitter=False,dodge=True)
plt.show()
#Using dodge and jitter
```

```python

sns.swarmplot(data=t,x='sex',y='age',hue='pclass',dodge=True)
plt.show()
#Using a swarmp plot
```

```python
sns.violinplot(data=t,x='sex',y='age',hue='survived',jitter=True,dodge=True)
plt.show()
#Using a violin plot for survival distribution of passengers
```

```python
sns.set(font_scale=1)
sns.swarmplot(data=t,x='sex',y='age',hue='pclass',dodge=True,color='grey')
sns.violinplot(data=t,x='sex',y='age',hue='pclass',jitter=False,dodge=True)
plt.show()
#Combing two different plots for better statistical analysis
```

```python
sns.set(font_scale=1)
sns.violinplot(data=t,x='pclass',y='age',hue='sex',dodge=True,split=True)
plt.show()
#Using the split parameter in violin plot
#Note:spli reuqires hue of only two categories not more
```

```python
sns.barplot(data=t,x='sex',y='age',hue='pclass',dodge=True)
plt.show()
#The bar plot has a black vertical line at the top that represents mean
```

```python
#Using joint and regression plots
```

```python

sns.jointplot(data=t,x='age',y='fare',kind='scatter')
plt.show()
#Using join plot for relation between fare and age
```

```python
#Trying a hex format here
sns.jointplot(data=t,x='age',y='fare',kind='hex')
plt.show()
```

```python
sns.jointplot(data=t,x='age',y='fare',kind='reg')
plt.show()
#The reg plot with ci of 95 proved mildly +ve for fare and age
```

```python
sns.lmplot(data=t,x='age',y='fare')
plt.show()# the following give thesame plot again here
```

```python
sns.lmplot(data=t,x='age',y='fare',hue='sex')
plt.show()
#Getting reg plots on hue splitting
```

```python
sns.lmplot(data=t,x='age',y='survived',hue='sex',col='sex',logistic=False)
plt.show()
#Studying corr between age and surv for sex
#Splitting on bases of col
```

```python
sns.lmplot(data=t,x='age',y='survived',col='pclass',logistic=True)
plt.show()
#studying corr between pclass with age and survived
#Hence we can see,lower the pclass more steap in the relation
```

```python
x=t.groupby(['sex','pclass']).survived.count().unstack(-1)
#Counting passengers wrt to plcss and sex
```

```python
#Creating Heat Maps of color representation
sns.heatmap(x,annot=True,fmt='d',cmap=None,vmax=None)
plt.show()
#These are great to study corr with colors
```

```python
sns.heatmap(x,annot=True,fmt='d',cmap='Reds',vmax=150)
plt.show()
```

```python
sns.set(font_scale=0.8)
sns.heatmap(t.corr(),annot=True,cmap='Reds',vmax=0.9)
plt.show()
#here we can see that pclass and fare have a -Ve corr
#sibs and parch have +ve corr
#fare and survived have +ve corr
```

```python
sns.set(font_scale=0.8)
sns.heatmap(t.groupby(['sex','pclass']).survived.mean().to_frame(),annot=True,cmap='Reds',vmax=0.9)
plt.show()
```
