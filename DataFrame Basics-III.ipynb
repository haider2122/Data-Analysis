```python
s=pd.read_csv('summer.csv')
```

```python
s.loc[(s.Discipline=='Swimming') & (s.Year>2007)].size
```

```python
s_01=s.loc[(s.Discipline=='Swimming') & (s.Year>2007)].copy()
```

```python
s_01.set_index('Athlete',inplace=True).head()
```

```python
x=s_01.index.value_counts().copy()
```

```python
s_01.insert(loc=5,column='Medals',value=x)
```

```python
s_01.sort_values(by='Medals',ascending=False)
```

```python
#Ranking players based on Medals
s_01.Medals.sort_values(ascending=False).rank(ascending=False,pct=False,method='min')
```
